```python
import os
import numpy as np

# 设置手势名称
gesture_name = "two"  # 可以替换为你想处理的任何手势
gesture_dir = f"gesture_dataset/{gesture_name}"

# 创建保存切割数据的文件夹
segment_dir = f"segment_data/{gesture_name}"
os.makedirs(segment_dir, exist_ok=True)

# 设置每个样本的帧数（例如 30 帧）
sequence_length = 30

# 获取所有的 .npy 文件
files = [f for f in os.listdir(gesture_dir) if f.endswith('_landmarks.npy')]

# 存储切割样本的计数器
sample_counter = 1

# 遍历每个手势数据文件
for file in files:
    file_path = os.path.join(gesture_dir, file)
    
    # 加载当前 .npy 文件的数据
    landmarks = np.load(file_path)  # shape: (num_frames, 21, 3)
    
    # 获取总帧数
    num_frames = landmarks.shape[0]

    # 切割数据，生成多个小样本，每个样本包含 sequence_length 帧
    for i in range(num_frames - sequence_length + 1):
        sample = landmarks[i:i + sequence_length]  # 获取30帧数据
        
        # 保存切割数据为新的 .npy 文件
        save_name = f"sample_{sample_counter}.npy"
        save_path = os.path.join(segment_dir, save_name)
        np.save(save_path, sample)
        
        print(f"保存样本：{save_name}")
        
        sample_counter += 1  # 更新样本计数器

print("数据切割完成，所有样本已保存。")
```
