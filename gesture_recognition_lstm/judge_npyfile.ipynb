```python
import numpy as np
import matplotlib.pyplot as plt

# 加载数据
landmarks = np.load("/Users/ruiyuhan/Desktop/Advanced/gesture_recognition_lstm/train_data/fist/fist_398.npy")

# 取前10帧数据进行可视化
for i in range(10):
    frame_landmarks = landmarks[i]  # 获取第 i 帧的关键点

    # 提取 x, y 坐标
    x = frame_landmarks[:, 0]
    y = frame_landmarks[:, 1]

    # 绘制关键点
    plt.figure(figsize=(5, 5))
    plt.scatter(x, y, c='blue', marker='o', label='Hand Landmarks')
    plt.title(f"Frame {i+1} Hand Landmarks")
    plt.xlabel('X Coordinate')
    plt.ylabel('Y Coordinate')
    plt.gca().invert_yaxis()  # 反转Y轴，因为OpenCV的坐标系和图像坐标系不同
    plt.legend()
    plt.show()
```
